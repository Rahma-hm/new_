```python
last_week=str(DT.date.today() - DT.timedelta(days=7))+'T00:00:00.0Z'
keyword = input("keyword :")
def get_videos():
    req = auth().search().list(q=keyword, part='snippet', type='video',publishedAfter=last_week, maxResults=50, pageToken=None)
    res = req.execute()
    return res
    
```

```python
def keyword_with_space(keyword):
    keyword_with_space = keyword.replace(" ", "_")
    return keyword_with_space
```

```python
def save_to_hdfs():
    
    client = InsecureClient('http://localhost:9870/',user='youtube')
    file_name =keyword_with_space(keyword)+'_youtube'+str(DT.date.today())+'.json'
    text_to_write= "\n".join([json.dumps( get_videos())])
    with client.write(file_name, encoding = 'utf-8') as writer:
        writer.write(text_to_write)
    
```
